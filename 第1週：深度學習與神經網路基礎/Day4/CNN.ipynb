實作一個優化過的 CNN，並使用池化和 Dropout 技術來提升模型的性能。

這次我們依然會使用 **TensorFlow** 和 **PyTorch** 進行實作，並展示如何將池化層和 Dropout 應用到模型中。

### **TensorFlow 實作 CNN**


#### **優化方法說明：**
- 我們在每個卷積層後加入了**最大池化層（Max Pooling Layer）**來減少特徵圖的大小。
- 加入 **Dropout 層**，在卷積層和全連接層後防止過擬合，特別是在全連接層使用了 $50\%$ 的 Dropout。
- 使用 **Adam 優化器**來加速訓練過程，並提高模型的收斂速度。

```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import mnist

# 下載並預處理 MNIST 資料集
# MNIST 是一個手寫數字的資料集，包含 60,000 張訓練圖像和 10,000 張測試圖像
(x_train, y_train), (x_test, y_test) = mnist.load_data()
# 將訓練和測試圖像數據重塑為 (28, 28, 1) 並正規化到 0-1 範圍
x_train = x_train.reshape((x_train.shape[0], 28, 28, 1)) / 255.0  # 增加一個通道維度，並進行正規化
x_test = x_test.reshape((x_test.shape[0], 28, 28, 1)) / 255.0

# 建立 CNN 模型，加入 Dropout 和池化
model = models.Sequential([
    # 第一個卷積層，使用 32 個 3x3 的卷積核，激活函數為 ReLU，輸入形狀為 (28, 28, 1)
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    # 最大池化層，池化窗口大小為 2x2
    layers.MaxPooling2D((2, 2)),
    # Dropout 層，隨機丟棄 25% 的神經元，防止過擬合
    layers.Dropout(0.25),
    # 第二個卷積層，使用 64 個 3x3 的卷積核，激活函數為 ReLU
    layers.Conv2D(64, (3, 3), activation='relu'),
    # 最大池化層，池化窗口大小為 2x2
    layers.MaxPooling2D((2, 2)),
    # 再次 Dropout 層，隨機丟棄 25% 的神經元，防止過擬合
    layers.Dropout(0.25),
    # 展開層，將多維數據展開為一維
    layers.Flatten(),
    # 全連接層，包含 128 個神經元，激活函數為 ReLU
    layers.Dense(128, activation='relu'),
    # 大量 Dropout 層，隨機丟棄 50% 的神經元，防止過擬合
    layers.Dropout(0.5),
    # 輸出層，包含 10 個神經元，對應 10 個類別，激活函數為 softmax
    layers.Dense(10, activation='softmax')
])

# 編譯模型
# 使用 Adam 優化器，損失函數為 sparse_categorical_crossentropy，評估指標為準確率
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# 訓練模型
# 使用訓練數據進行訓練，訓練 10 個 epoch，每批次大小為 64
model.fit(x_train, y_train, epochs=10, batch_size=64)

# 評估模型
# 使用測試數據進行評估，並輸出測試準確率
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
print(f'Test accuracy: {test_acc}')
```

### **PyTorch 實作 CNN**


#### **優化方法說明：**
- 在每個卷積層後加入 **最大池化層（Max Pooling）** 來降低特徵圖的大小。
- 使用兩層 **Dropout** 防止過擬合，特別是在全連接層使用了 $50\%$ 的 Dropout。
- 使用 **Adam 優化器** 提高模型的收斂速度和效果。

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# 資料轉換與預處理
# 將圖像轉換為張量並標準化到 [-1, 1] 範圍
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,), (0.5,))])

# 下載並加載 MNIST 資料集
# MNIST 是一個手寫數字的資料集，包含 60,000 張訓練圖像和 10,000 張測試圖像
train_data = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
test_data = datasets.MNIST(root='./data', train=False, download=True, transform=transform)

# 使用 DataLoader 將資料集分批次加載
train_loader = DataLoader(train_data, batch_size=64, shuffle=True)
test_loader = DataLoader(test_data, batch_size=64, shuffle=False)

# 創建 CNN 模型，加入池化和 Dropout 層
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        # 第一個卷積層，輸入通道數為 1，輸出通道數為 32，卷積核大小為 3x3，步幅為 1
        self.conv1 = nn.Conv2d(1, 32, 3, 1)
        # 最大池化層，池化窗口大小為 2x2
        self.pool = nn.MaxPool2d(2, 2)
        # 第二個卷積層，輸出通道數為 64，卷積核大小為 3x3，步幅為 1
        self.conv2 = nn.Conv2d(32, 64, 3, 1)
        # Dropout 層，隨機丟棄 25% 的神經元，防止過擬合
        self.dropout1 = nn.Dropout(0.25)
        # 第二個 Dropout 層，隨機丟棄 50% 的神經元，防止過擬合
        self.dropout2 = nn.Dropout(0.5)
        # 全連接層，輸入大小為 5*5*64，輸出大小為 128
        self.fc1 = nn.Linear(5 * 5 * 64, 128)
        # 輸出層，輸出大小為 10，對應 10 個類別
        self.fc2 = nn.Linear(128, 10)
    
    def forward(self, x):
        # 前向傳播
        x = torch.relu(self.conv1(x))            # 第一個卷積層 + ReLU 激活
        x = self.pool(x)                         # 最大池化
        x = self.dropout1(x)                     # Dropout
        x = torch.relu(self.conv2(x))            # 第二個卷積層 + ReLU 激活
        x = self.pool(x)                         # 最大池化
        x = self.dropout1(x)                     # 再次 Dropout
        x = x.view(-1, 5 * 5 * 64)             # 展開為一維向量
        x = torch.relu(self.fc1(x))              # 全連接層 + ReLU 激活
        x = self.dropout2(x)                     # 大量 Dropout
        x = self.fc2(x)                          # 輸出層
        return torch.softmax(x, dim=1)           # 使用 softmax 進行分類

# 初始化模型、損失函數與優化器
model = CNN()
# 使用交叉熵損失函數
criterion = nn.CrossEntropyLoss()
# 使用 Adam 優化器，學習率為 0.001
optimizer = optim.Adam(model.parameters(), lr=0.001)

# 訓練模型
def train(model, train_loader, criterion, optimizer, epochs=10):
    model.train()  # 設置模型為訓練模式
    for epoch in range(epochs):
        running_loss = 0.0
        for images, labels in train_loader:
            optimizer.zero_grad()           # 梯度歸零
            outputs = model(images)         # 前向傳播
            loss = criterion(outputs, labels)  # 計算損失
            loss.backward()                 # 反向傳播
            optimizer.step()                # 更新權重
            running_loss += loss.item()     # 累加損失
        print(f'Epoch {epoch+1}, Loss: {running_loss/len(train_loader)}')  # 輸出每個 epoch 的平均損失

# 測試模型
def test(model, test_loader):
    model.eval()  # 設置模型為評估模式
    correct = 0
    total = 0
    with torch.no_grad():  # 禁用梯度計算
        for images, labels in test_loader:
            outputs = model(images)         # 前向傳播
            _, predicted = torch.max(outputs.data, 1)  # 獲取預測結果
            total += labels.size(0)         # 累加總樣本數
            correct += (predicted == labels).sum().item()  # 累加正確預測數
    print(f'Test Accuracy: {100 * correct / total}%')  # 輸出測試準確率

# 執行訓練與測試
train(model, train_loader, criterion, optimizer, epochs=10)
test(model, test_loader)
```

這兩個版本的實作展示了如何使用池化層來降低特徵圖尺寸，並通過 Dropout 技術來防止過擬合。這樣的模型在面對訓練集和測試集時應該能表現得更穩定。
